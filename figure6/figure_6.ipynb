```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress
import os
```

```python
#Import the file with 'context' manually added NOTE: Manually added data needed to be cross referenced
#from two publically available files in the repository
data3 = pd.read_csv(r"mec15689-sup-0002-table-s7-context.csv")

count_CG = data3['Context'].value_counts().get('CG', 0)
count_CHG = data3['Context'].value_counts().get('CHG', 0)
count_CHH = data3['Context'].value_counts().get('CHH', 0)

#Count the number of CG occurences for easier visual processing
print("Number of total'CG' occurrences:", count_CG)

#Count the number of CHG occurences for easier visual processing
print("Number of total'CHG' occurrences:", count_CHG)

#Count the number of CHH occurences for easier visual processing
print("Number of total'CHH' occurrences:", count_CHH)
```

```python
#Create a dataframe for CG 'context'
CG_df = data3[data3['Context'].str.contains('CG', na=False)]

#Filtering for DNA Transposons only from the 'TE funciton' column
DNAtrans_CG_df = CG_df[CG_df['TE function'].str.contains('DNAtrans')]

#Counting the number of CG for visual processing, should be different than the above number
count_CG = DNAtrans_CG_df['Context'].value_counts().get('CG', 0)
print("Number of 'CG' occurrences:", count_CG)
```

```python
#create a dataframe for CHG 'context'
CHG_df = data3[data3['Context'].str.contains('CHG', na=False)]

#Filtering for DNA Transposons only from the 'TE function' column
DNAtrans_CHG_df = CHG_df[CHG_df['TE function'].str.contains('DNAtrans')]

#Counting the number of CHG for visual processing, should be different than the above number
count_CHG = DNAtrans_CHG_df['Context'].value_counts().get('CHG', 0)
print("Number of 'CHG' occurrences:", count_CHG)
```

```python
#create a dataframe for CHH 'context'
CHH_df = data3[data3['Context'].str.contains('CHH', na=False)]

#Filtering for DNA Transposons only from the 'TE function' column
DNAtrans_CHH_df = CHH_df[CHH_df['TE function'].str.contains('DNAtrans')]

#Counting the number of CHH for visual processing, should be different than the above number
count_CHH = DNAtrans_CHH_df['Context'].value_counts().get('CHH', 0)
print("Number of 'CHH' occurrences:", count_CHH)
```

```python
# Plot density of methylation from transposons for each context using seaborn
sns.kdeplot(DNAtrans_CG_df['MethDiffProp'], shade=False, color='orange', label='CG')
sns.kdeplot(DNAtrans_CHH_df['MethDiffProp'], shade=False, color='blue', label='CHH')
sns.kdeplot(DNAtrans_CHG_df['MethDiffProp'], shade=False, color='green', label='CHG')

# Add a vertical dashed line at x = 0
plt.axvline(0, color='black', linestyle='--')

# Set labels and title
plt.xlabel('Arbitrary Units')
plt.ylabel('DMC Density')
plt.title('DNA Transposons')

# Show legend
plt.legend()

#Save the plot
plt.savefig('DNA_Transposons_Density.png')

# Show the plot
plt.show()

```
